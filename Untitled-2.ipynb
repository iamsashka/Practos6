Создаем новую фигуру (график) с заданным размером. Строим линейный график, где по оси x отображаются последние 25 значений столбца 'Set_ID', а по оси y - последние 25 значений столбца 'Year'. Опция label создает метку для легенды, color="purple" устанавливает цвет графика, marker='o' добавляет круглые маркеры, и markersize=7 устанавливает размер маркеров.Устанавливаем заголовок для графика. Устанавливаем подписи осей x и y соответственно.Добавляем легенду, которая в данном случае будет отображать метку 'Разговорчивость'.
Добавляем сетку на график для улучшения читаемости.
Таким образом, этот код создает линейный график, отражающий статистику разговорчивости последних 5 треков в вашем датасете. Маркеры 'o' позволяют визуально выделить отдельные точки, и сетка делает график более наглядным.

`plt.figure(figsize=(20,10))` создаёт новую фигуру (график) с размером 20 на 10 дюймов.

`plt.plot(sets['Set_ID'][1000:1500],sets['Year'][1000:1500], label = 'Разговорчивость', color = "purple", marker='o', markersize=7)` строит линейный график, где по оси x отображаются значения столбца 'Set_ID' для индексов с 1000 по 1500, а по оси y - значения столбца 'Year' для тех же индексов. Опция `label` создаёт метку для легенды графика, `color="purple"` задаёт цвет графика, `marker='o'` добавляет круглые маркеры для каждой точки на графике, `markersize=7` устанавливает размер маркеров.

```python
plt.figure(figsize=(20,10))
plt.plot(sets['Set_ID'][1000:1500],sets['Year'][1000:1500], label = 'Разговорчивость', color = "purple", marker='o', markersize=7)
plt.title('Статистика года')
plt.xlabel(' ID Названиия')
plt.ylabel('год')
plt.legend()
plt.grid()
plt.show()
```

Создаем новую фигуру (график) с заданным размером. Строим линейный график, где по оси x отображаются последние 25 значений столбца 'Set_ID', а по оси y - последние 25 значений столбца 'Pieces'. Добавляем столбчатую диаграмму поверх линейного графика. Опция label создает метку для легенды, а color='pink' устанавливает цвет столбцов.Устанавливаем заголовок для графика. Устанавливаем подписи осей x и y соответственно.
Добавляем легенду, которая будет содержать две метки: 'Фрагменты' для линейного графика и 'Фрагменты столбцы' для столбчатой диаграммы.
Таким образом, мы получаем гибридный график, который позволяет визуализировать популярность треков как с использованием линий, так и столбцов.

```python
plt.figure(figsize=(20,10))
plt.plot(sets['Set_ID'].head(25),sets['Pieces'].head(25), label = 'ID фрагмента', color = "purple")
plt.bar(sets['Set_ID'].head(25),sets['Pieces'].head(25), label = 'ID фрагмента столбцы', color = 'pink')
plt.title('Статистика части наборов')
plt.xlabel('Названиие')
plt.ylabel('Фрагменты')
plt.legend()
plt.show()
```

Код создает столбчатую диаграмму для отображения года последних 25 наборов из DataFrame sets. Ось x представляет названия, ось y - год. Цвет столбцов - красный.

```python
plt.figure(figsize=(20,10))
plt.bar(sets['Set_ID'].head(25),sets['Pieces'].head(25), label = 'ID фрагмента', color = 'red')
plt.title('Статистика части наборов')
plt.xlabel('Названиие')
plt.ylabel('Фрагменты')
plt.legend()
plt.show()
```

Код визуализирует ID последних 15  фрагментов с использованием линейного графика и столбчатой диаграммы для отображения информации об исполнителях. Также есть настройки для размера фигуры, цвета, маркеров и других параметров

```python
plt.figure(figsize=(20,10))
plt.plot(sets['Set_ID'].head(15),sets['Pieces'].head(15), label = 'ID фрагмента', color = "purple", markersize=7, marker='o')
plt.bar(sets['Set_ID'].head(15),sets['Pieces'].head(15), label = 'Исполнитель', color = 'pink')
plt.title('Статистика части наборов')
plt.xlabel('Названиие')
plt.ylabel('Фрагменты')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.bar(sets['Set_ID'].head(25),sets['Pieces'].head(25), label = 'Популярность трека', color = 'red')
plt.bar(sets['Set_ID'].head(25),sets['Name'].head(25), label = 'Название', color = 'purple')
plt.legend()
plt.xlabel('Статистика части наборов')
plt.ylabel('Название')
plt.show()
```

```python
Set_ID = sets['Set_ID'].value_counts()
plt.pie(Set_ID, labels=Set_ID.index, autopct='%1.0f%%', colors = ['orange','pink'])
plt.title('Год')
plt.show()
```

```python
key = sets['Owned'].value_counts()
valid_keys = key[key.index <= 5]
labels = valid_keys.index
plt.pie(valid_keys, labels=labels, autopct='%1.1f%%', colors = ['orange','pink','red','purple','blue','green'])
plt.title('Ключ')
plt.show()
```
